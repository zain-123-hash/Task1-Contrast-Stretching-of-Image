# Task
Calculate and apply a linear contrast stretching function with a contrast enhancement factor of 2 to the image "/content/boy in street.jpg", mapping the original pixel value range of 10-70 to a new range of 0-255. Display both the original and the processed images.

## Load the image

### Subtask:
Load the image from "/content/boy in street.jpg" into a suitable data structure, like a NumPy array.


**Reasoning**:
Import the necessary library and load the image into a NumPy array.



```python
import cv2

# Load the image from the specified path
image_path = '/content/boy in street.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

**Reasoning**:
Display the loaded image to verify it was loaded correctly.



```python
from google.colab.patches import cv2_imshow

# Display the loaded image
cv2_imshow(image)
```

## Define the contrast stretching function

### Subtask:
Create a Python function that takes a pixel value, the original minimum and maximum values, the new minimum and maximum values, and a contrast enhancement factor as input, and returns the new pixel value after applying the linear contrast stretching with the given factor.


**Reasoning**:
Define the `linear_contrast_stretch` function as described in the instructions.



```python
def linear_contrast_stretch(pixel_value, original_min, original_max, new_min, new_max, contrast_factor):
    """
    Applies linear contrast stretching with a contrast enhancement factor.

    Args:
        pixel_value: The original pixel value.
        original_min: The minimum value of the original range.
        original_max: The maximum value of the original range.
        new_min: The minimum value of the new range.
        new_max: The maximum value of the new range.
        contrast_factor: The factor to enhance contrast.

    Returns:
        The new pixel value after applying contrast stretching.
    """
    if pixel_value < original_min:
        return new_min
    elif pixel_value > original_max:
        return new_max
    else:
        new_pixel_value = new_min + (new_max - new_min) * ((pixel_value - original_min) / (original_max - original_min)) * contrast_factor
        return new_pixel_value
```

## Apply the function

### Subtask:
Apply the defined function to each pixel of the loaded image to create the contrast-stretched image.


**Reasoning**:
Apply the defined `linear_contrast_stretch` function to each pixel of the image to create the contrast-stretched image.



```python
import numpy as np

# Create a new NumPy array to store the contrast-stretched image
stretched_image = np.zeros_like(image)

# Iterate through each pixel and apply the contrast stretching function
original_min = 10
original_max = 70
new_min = 0
new_max = 255
contrast_factor = 2

for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        stretched_image[i, j] = linear_contrast_stretch(image[i, j], original_min, original_max, new_min, new_max, contrast_factor)
```

## Display the results

### Subtask:
Display both the original and the contrast-stretched images for comparison.


**Reasoning**:
Import the necessary function for displaying images in a Colab environment and then display both the original and contrast-stretched images with descriptive titles.



```python
from google.colab.patches import cv2_imshow

# Display the original image
print("Original Image:")
cv2_imshow(image)

# Display the contrast-stretched image
print("Contrast-Stretched Image:")
cv2_imshow(stretched_image)
```
